```python
df = spark.table("zomato_reviews_cluster.default.clean_zomato_reviews")
df.show(5)
df.printSchema()
```

```python
## Basic Stats
df.describe(["rating", "helpful", "review_length_words"]).show()
```

```python
## Avg rating by Month
from pyspark.sql.functions import avg

df_avg_rating_by_month = (
    df.groupBy("review_month")
      .agg(avg("rating").alias("avg_rating"))
      .orderBy("review_month")
)

df_avg_rating_by_month.show()
```

```python
## Avg helpful VOTES by Rating
df_helpful_by_rating = (
    df.groupBy("rating")
      .agg(avg("helpful").alias("avg_helpful"))
      .orderBy("rating")
)

df_helpful_by_rating.show()
```

```python
## Avg review length by rating category
df_length_by_rating_cat = (
    df.groupBy("rating_category")
      .agg(avg("review_length_words").alias("avg_review_length_words"))
      .orderBy("rating_category")
)

df_length_by_rating_cat.show()
```

```python
df = spark.sql(
  """
  SELECT
    review_year,
    AVG(rating) AS avg_rating
  FROM zomato_reviews_cluster.default.clean_zomato_reviews
  GROUP BY review_year
  ORDER BY review_year
  """
)
display(df)
```

```python
df = spark.sql(
  """
  SELECT
    rating,
    AVG(helpful) AS avg_helpful
  FROM zomato_reviews_cluster.default.clean_zomato_reviews
  GROUP BY rating
  ORDER BY rating
  """
)
display(df)
```
