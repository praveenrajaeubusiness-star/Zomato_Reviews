```python
from pyspark.sql.functions import avg

df = spark.table("zomato_reviews_cluster.default.clean_zomato_reviews")
df.show(5)
```

```python
## Avg rating by Month
dash_avg_rating_by_month = (
    df.groupBy("review_month")
      .agg(avg("rating").alias("avg_rating"))
)

dash_avg_rating_by_month.write.mode("overwrite").saveAsTable(
    "zomato_reviews_cluster.default.dash_avg_rating_by_month"

)
```

```python
## Average helpful votes by rating
dash_helpful_by_rating = (
    df.groupBy("rating")
      .agg(avg("helpful").alias("avg_helpful"))
)

dash_helpful_by_rating.write.mode("overwrite").saveAsTable(
    "zomato_reviews_cluster.default.dash_helpful_by_rating"
)
```

```python
## Average review length by rating category
dash_length_by_rating_cat = (
    df.groupBy("rating_category")
      .agg(avg("review_length_words").alias("avg_review_length_words"))
)

dash_length_by_rating_cat.write.mode("overwrite").saveAsTable(
    "zomato_reviews_cluster.default.dash_length_by_rating_cat"
)
```
